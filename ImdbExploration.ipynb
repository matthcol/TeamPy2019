### Steve McQueen (a lot of)

```python
# search for Steve McQueen and some others
steves = ia.search_person('Steve McQueen')
steves
```

```python
# keep only the true Steve McQueen (s)
steveMcQueenList = [ s for s in steves 
                    if re.fullmatch(r'Steve McQueen', s['name'])]
steveMcQueenList
```

```python
# update true Steve McQueen (s) to retreive Filmography
for steve in steveMcQueenList:
    ia.update(steve)
```

```python
# display filmography as an actor or a director for each steve

def displayFilmographySection(filmography, section):
    """" searching the right section (dict) in the filmography """
    sectionMovies = None
    for filmo in filmography:
        if section in filmo:
            # found the right categorie
            sectionMovies = filmo
            break
    # print movies of the category if found
    if sectionMovies is not None:
        print(f'\t- as {section}:')
        for movie in sectionMovies[section]:
            title = movie['title']
            # year may be missing
            year = movie['year'] if 'year' in movie else None
            print(f'\t\t* {title} ({year})')
    
for steve in steveMcQueenList:
    # ia.update(steve) # already done before
    name = steve['name']
    birthdate = steve['birth date'] if 'birth date' in steve else None
    print(f'. {name} ({birthdate}):')
    filmography = steve['filmography']
    displayFilmographySection(filmography, 'actor')
    displayFilmographySection(filmography, 'director')
    
```
